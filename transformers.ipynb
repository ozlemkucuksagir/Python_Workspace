```python
from transformers import pipeline
```

```python
cls=pipeline("sentiment-analysis")
```

```python
text="I am feeled lucky"
```

```python
cls(text)
```

```python
import pandas as pd
output=cls(text)
pd.DataFrame(output)
```

```python
cls=pipeline("zero-shot-classification")
text="this tutorial is a about huggingface"
```

```python
labels=["tech","education","business"]
```

```python
output=cls(text,labels)
pd.DataFrame(output)
```

```python
generator=pipeline("text-generation")
```

```python
prompt="this tutorial will walk you throught how to"
output=generator(prompt)
```

```python
output
```

```python
output=generator(prompt,max_length=100,num_return_sequences=1)
print(output,"generated-text")
```
